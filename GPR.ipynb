```
import pandas as pd
import matplotlib.pyplot as plt
```

```
dsf = 'https://www2.bc.edu/matteo-iacoviello/gpr_files/gpr_web_latest.xlsx' #Latest GPR Index data
ds = pd.read_excel(dsf) #DF for xlsx
```

```
ds_sub=ds[["Date","GPR"]] #Filter for Date and GPR
df=ds_sub[ds_sub.GPR >= 0] #Filter out footer
#df['GPR'] = df['GPR'].replace(',','.', regex=True).astype(float)
plt.plot(df['Date'], df['GPR'])
plt.xticks(rotation='vertical')
```
